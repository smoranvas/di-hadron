```python
import numpy as np, pandas as pd, ROOT, matplotlib.pyplot as plt
```

```python
# Check that I'm getting the right results from the BeAGLE rivet pipeline
```

```python
tars="D C Fe Pb".split()
data={i:{} for i in range(6)}
for tar in tars:
    print(tar)
    deg=np.pi/180
    dys=[]
    dphis=[]
    npiplead=0
    for i in range(1,101):
        print(i)
        with open(f"../hepmc/e{tar}_{i}.hepmc", "r") as f:
            line=f.readline()
            newEvent=False
            ln=0
            
            pips=[]
            prots=[]
            while line !="":
                ln+=1
                line=f.readline()
                #print(line)
                #if ln==100:
                #    break;
                if line.startswith("U GEV MM") and len(pips)>0:
                    #print("new event")
                    q=ebeam-e
                    Q2=-q.M2()
                    W=(q+pbeam).M()
                    nu=q.E()
                    qph=q.Phi()
                    qth=q.Theta()
                    if W>2 and Q2>1 and nu>2.3 and nu<4.2:
                        for pip in pips:
                            
                            if not(pip.E()>nu/2 and pip.Vect().Angle(ebeam.Vect())>10*deg):
                                continue
                            pip.RotateZ(-qph)
                            pip.RotateY(-qth)
                            if pip.Pt()<0.07:
                                continue
                            npiplead+=1
                            for prot in prots:
                                if prot.P()<0.35:
                                    continue
                                if not(prot.Vect().Angle(ebeam.Vect())>10*deg):
                                    continue
                                if pip.Pt()<0.07:
                                    continue
                                prot.RotateZ(-qph)
                                prot.RotateY(-qth)
                                dphi=pip.Phi()-prot.Phi()
                                
                                dys.append(pip.Rapidity()-prot.Rapidity())
                                dphis.append(dphi)
                                        
                if line.startswith("U GEV MM"):    
                    ebeam=None
                    pbeam=None
                    pips=[]
                    prots=[]
                if line.startswith("P 1 0 11"):
                    ebeam=ROOT.TLorentzVector(*[float(a) for a in line.split()[4:8]])
                if line.startswith("P 2 0 2212"):
                    pbeam=ROOT.TLorentzVector(*[float(a) for a in line.split()[4:8]])
                        
                split=line.split()
                if(len(split)<4):
                    continue
                if split[3]=="11":
                    e=ROOT.TLorentzVector(*[float(a) for a in line.split()[4:8]])
                elif split[3]=="211":
                    #print("pip")
                    pips.append(ROOT.TLorentzVector(*[float(a) for a in line.split()[4:8]]))
                elif split[3]=="2212":
                    prots.append(ROOT.TLorentzVector(*[float(a) for a in line.split()[4:8]]))
    dphis=np.array(dphis)
    dphis=np.abs(dphis+(dphis>np.pi)*-2*np.pi+(dphis<-np.pi)*2*np.pi)
    h,x,y,_=plt.hist2d(dys,dphis, bins=(6, 8), range=((0,3),(0,np.pi)))

    if tar=='D':
        dY=.5
        dphi=np.pi/8
        norm=np.sum(h)*dY*dphi/npiplead
    C=h/npiplead/norm
    dCstat=np.sqrt(h+(h==0))/npiplead/norm
    phi=(y[1:]+y[:-1])/2
    phi_low=y[:-1]
    phi_up=y[1:]
    if tar=="D":
        C_D=C
        dCstat_D=dCstat
    R=C/C_D
    dRstat=R*np.hypot(dCstat/C,dCstat_D/C_D)
    for k in range(6):
        print(C)
        d=dict(C=C[k], dCstat=dCstat[k], 
                                            phi=phi,phi_low=phi_low, phi_up=phi_up,
                                                R=R[k],
                                                dRstat=dRstat[k])
        print({a:len(d[a]) for a in d})
        data[k][tar] = pd.DataFrame(d)
            
#U GEV MM
```
