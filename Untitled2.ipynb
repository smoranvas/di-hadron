```python

edges = R2h_module.binEdges_mpair
for a,col in zip('D C Fe Pb'.split(),"krgb"):
    df = dfs_data[f'{a}_2pi']
    x = []
    y1 = []
    y2 = []
    y3 = []
    for i in range(len(edges)-1):
        dfq = df.query(f'pair_mass>{edges[i]} and pair_mass< {edges[i+1]}')
        y1.append(np.mean(dfq.h1_cm_pt))
        y2.append(np.mean(dfq.h1_cm_pt)+np.std(dfq.h1_cm_pt))
        y3.append(np.mean(dfq.h1_cm_pt)-np.std(dfq.h1_cm_pt))
        x.append((edges[i]+edges[i+1])/2)
    plt.plot(x,y1, label=a, linestyle='-', color=col)
    plt.plot(x,y2, label=None, linestyle='--', color=col, alpha=0.5)
    plt.plot(x,y3, label=None, linestyle='--', color=col, alpha=0.5)
plt.gca().set_xlabel('$m_{\\mathrm{pair}}$')
plt.gca().set_ylabel('$<p_T^{\\mathrm{pair}}>$')
plt.legend()
plt.savefig("images/ptpair_vs_mpair.png")
    #plt.plot(x,y2, label=a+", max", linestyle='--', color=col)
```
