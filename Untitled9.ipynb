```python
import time,os
from matplotlib.offsetbox import AnchoredText
import sys,pandas as pd, matplotlib , matplotlib.pyplot as plt, matplotlib.lines, numpy as np, math, pylab

import uproot3
%matplotlib inline
```

```python
dfs_pair={}
dfs_trigger={}
#get GiBUU
import uproot

def getdatainChunks(filename,treename):
    dataframe =pd.DataFrame()
    file = uproot.open(filename) 
    tree=file[treename]
    df=tree.arrays(filter_name='*',library='pd')
    return df

pathGIBUU="./"
for target in ['D','C','Fe','Pb']:
    for suffix in "_tweak_harut",:
        #GiBUU
        print (target+suffix)
        print (' trigger')
        dfs_trigger['GiBUU_%s'%target+suffix]= getdatainChunks('%sGiBUU_Pairs_%s.root'%(pathGIBUU,target+suffix), 
                                                               '%s_trigger'%target)
        print (' pairs')
        dfs_pair['GiBUU_%s'%target+suffix]= getdatainChunks('%sGiBUU_Pairs_%s.root'%(pathGIBUU,target+suffix),target)
print (dfs_trigger.keys())
```

```python
import matplotlib as mpl
df=dfs_pair['GiBUU_D_tweak_harut'].query("h1_cm_pt>0.25 and h2_cm_pt>0.25")
h2d=plt.hist2d(df.eval("h1_cm_ph"), df.eval("dphi"), bins=180, norm = mpl.colors.LogNorm())
plt.xlabel("$\phi_1$ [rad]")
plt.ylabel("$|\Delta\phi|$ [rad]")
#cb = h2d[0].colorbar()
#cb[0].set_zlabel("# of GiBUU events")
plt.show()
```
